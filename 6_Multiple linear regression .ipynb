```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\HP\Desktop\Machine Learning\mul_regression.csv")
```

```python
data.head()
```

```python
x = data.iloc[:, :-1]
x.head()
```

```python
y = data.iloc[:,-1].values
y
```

```python
res = pd.get_dummies(x['State'], drop_first = True)   # get_dummies will convert categorical data to numerical
res.head()                                # get_dummies performs one hot encoding

#drop_first = True , to remove one column. i.e., to remove multicollinearity
```

```python
x = x.drop(['State'], axis = 1)   # drop is a function used to drop either rows or columns by giving specific numbers

#axis = 0 means drop rowwise and axis =1 means drop columnwise
```

```python
x.head()
```

```python
x = pd.concat((x, res), axis =1)
x.head()
```

```python
x = x.values  # 2-dim array
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train, y_test = train_test_split(x, y , test_size = 1/3,
                                                 random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_   # since there are 5 (p=5) independent columns we get 5 values for reg coefficient
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2= r2_score(y_test, y_pred)
```

```python
len(x_test) #n=17
```

```python
adj_r2 = 1-((1-r2) * (16/11))               ##1-((1-r2) * (17-1/17-5-1))
```
